```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1


clmo.png

```python
#Manage data for variable 1
df['CLOSENESS_MOTHER'] = df['H4WP24']
df['CLOSENESS_MOTHER'].value_counts()
```

```python
df['CLOSENESS_MOTHER'] = df['CLOSENESS_MOTHER'].loc[ df['CLOSENESS_MOTHER'] !=' ' ]
df['CLOSENESS_MOTHER'] = df['CLOSENESS_MOTHER'].loc[ df['CLOSENESS_MOTHER'] !='7' ]
df['CLOSENESS_MOTHER'].value_counts()
```

```python
df['CLOSENESS_MOTHER'].replace(['1', '2', '3', '4', '5'] , ['Not At All Close', 'Not Very Close', 'Somewhat Close', 'Quite Close', 'Very Close'], inplace = True)
df['CLOSENESS_MOTHER'].value_counts()
```

```python
rp.summary_cat(df['CLOSENESS_MOTHER'])
```

```python
#Graphs and tables or statistics
```

```python
plt.title("How close do you feel to your (mother figure)?")
explode = (0.1, 0.1, 0.1, 0.5, 0.5)
df['CLOSENESS_MOTHER'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['palevioletred', 'crimson', 'pink', 'deeppink', 'orchid'])
```

```python
plt.xlabel("Closeness")
plt.title("How close do you feel to your (mother figure)?")
df['CLOSENESS_MOTHER'].value_counts(normalize = True).plot(kind ='barh', color = 'palevioletred', width = 0.2, alpha = 0.5)
```

verba.png

```python
#Manage data for variable 1
df['H4MA1'].value_counts()
```

```python
df['VERBAL_ABUSE'] = df['H4MA1']
df['VERBAL_ABUSE'].value_counts()
```

```python
df['VERBAL_ABUSE'] = df['VERBAL_ABUSE'].loc[ df['VERBAL_ABUSE'] !=' ' ]
df['VERBAL_ABUSE'] = df['VERBAL_ABUSE'].loc[ df['VERBAL_ABUSE'] !='96' ]
df['VERBAL_ABUSE'] = df['VERBAL_ABUSE'].loc[ df['VERBAL_ABUSE'] !='98' ]
df['VERBAL_ABUSE'].value_counts()
```

```python
df['VERBAL_ABUSE'].replace(['1', '2', '3', '4', '5', '6'] , ['1', '2', '3-5', '6-10', '10+', 'Never'], inplace = True)
df['VERBAL_ABUSE'].value_counts()
```

```python
rp.summary_cat(df['VERBAL_ABUSE'])
```

```python
#Graphs and tables or statistics
```

```python
plt.title("Before your 18th birthday, how often did a parent or other adult caregiver say things that really hurt your feelings or made you feel like you were not wanted or loved?")
explode = (0.1, 0.1, 0.1, 0.1, 0.4)
df['VERBAL_ABUSE'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['powderblue', 'skyblue', 'royalblue', 'cadetblue', 'steelblue', 'darkblue'])
```

```python
plt.xlabel("Verbal Abuse")
plt.title("Before your 18th birthday, how often did a parent or other adult caregiver say things that really hurt your feelings or made you feel like you were not wanted or loved?")
df['VERBAL_ABUSE'].value_counts(normalize = True).plot(kind ='barh', color = 'royalblue', width = 0.2, alpha = 0.5)
```
