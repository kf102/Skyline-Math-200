Research Question #1 : Is there an association between if you have recently hospitalized someone recently and how angry you are?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1

ang.png

```python
#Manage data for variable 1
df['H4PE8'].value_counts()
```

```python
df['ANGER'] = df['H4PE8']
df['ANGER'].value_counts()
```

```python
df['ANGER'] = df['ANGER'].loc[ df['ANGER'] !=' ' ]
df['ANGER'] = df['ANGER'].loc[ df['ANGER'] !='6' ]
df['ANGER'] = df['ANGER'].loc[ df['ANGER'] !='8' ]
df['ANGER'].value_counts()
```

```python
df['ANGER'].replace(['1', '2', '3', '4', '5'] , ['Strongly Agree', 'Agree', 'Neither Agree Nor Disagree', 'Disagree', 'Strongly Disagree'], inplace = True)
df['ANGER'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python
rp.summary_cat(df['ANGER'])
```

```python
plt.title("I get angry easily.")
explode = (0.1, 0.1, 0.1, 0.1, 0.4)
df['ANGER'].value_counts().plot(kind ='pie', autopct='%1.1f%%', explode=explode, colors=['darkgreen', 'seagreen', 'green', 'yellowgreen', 'olivedrab'])
```

```python
plt.xlabel("Anger Levels")
plt.title("I get angry easily.")
df['ANGER'].value_counts(normalize = True).plot(kind ='barh', color = 'darkgreen', width = 0.2, alpha = 0.5)
```

hosp.png

```python
#Manage data for variable 1
df['H4DS12'].value_counts()
```

```python
df['HOSPITALIZATION'] = df['H4DS12']
df['H4DS12'].value_counts()
```

```python
df['HOSPITALIZATION'] = df['HOSPITALIZATION'].loc[ df['HOSPITALIZATION'] !=' ' ]
df['HOSPITALIZATION'] = df['HOSPITALIZATION'].loc[ df['HOSPITALIZATION'] !='7' ]
df['HOSPITALIZATION'].value_counts()
```

```python
df['HOSPITALIZATION'].replace(['0', '1', '2', '3'] , ['Never', '1-2', '3-4', '5+'], inplace = True)
df['HOSPITALIZATION'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python
rp.summary_cat(df['HOSPITALIZATION'])
```

```python
plt.title("In the past 12 months, how often did you hurt someone badly enough in a physical fight that he or she needed care from a doctor or nurse?")
explode = (0.1, 0.1, 0.1, 0.4)
df['HOSPITALIZATION'].value_counts().plot(kind ='pie', autopct='%1.1f%%', explode=explode, colors=['maroon', 'firebrick', 'salmon', 'tomato'])
```

```python
plt.xlabel("Hospitalization in the Past Year")
plt.title("In the past 12 months, how often did you hurt someone badly enough in a physical fight that he or she needed care from a doctor or nurse?")
df['HOSPITALIZATION'].value_counts(normalize = True).plot(kind ='barh', color = 'maroon', width = 0.2, alpha = 0.5)
```
